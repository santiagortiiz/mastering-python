## Match Object Methods

```python
import re
```

### Dataset

```python
text = 'FORMULA MEDICA  Fecha de Atencin: 2021-05-05  Sede: UT VIVA 1A SEDE MEDELLIN Direccin: Cl 9C Sur HF 50FF - 116 Telfono: 6042220  Paciente: ROBERTO EPIFANIO CEBALLOS RIVERA ID: 535363  Contrato: UT VIVA 1A SEDE MEDELLIN Plan: CONTRIBUTIVO Semanas: 702 Rango: 1  Tipo de Usuario: COTIZANTE Sede Afiliado: UT VIVA 1A SEDE MEDELLIN'
text
```

### Search

```python
# Pattern design
pattern = 'paciente:?\s?([\w\s]*)\sID:?\s?(\d+)\s*'
print(f'Pattern: {pattern}')

# Browse
m = re.search(pattern, text, re.IGNORECASE)
print(m)
print(f'Bool: {bool(m)}\n\nBrowse:')
if (m.groups()):
    for i in m.groups():
        print(i)
```

### Split

```python
# Pattern design
pattern = 'paciente:?\s?([\w\s]*)\sID:?\s?(\d+)\s'
print(f'Pattern: {pattern}')

# Browse
m = re.split(pattern, text, re.IGNORECASE)
print(f'Browse: \n{m}')
```

### FindAll

```python
# Pattern design
pattern = 'paciente:?\s?([\w\s]*)\sID:?\s?(\d+)\s'
print(f'Pattern: {pattern}')

# Browse
m = re.findall(pattern, text, re.IGNORECASE)
print(f'Browse: \n{m}')
```

```python
type(m[0])
```
