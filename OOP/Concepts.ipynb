# Access Specifiers

```python
class A:
    
    def __init__(self, a, b):
        self._protected = a
        self.__private = b
```

```python
a = A('protected attr', 'private attr')
```

```python
a._protected = 'protected updated'
a._protected
```

```python
a.__private
```

# Get & Set Methods

```python
class House:

    def __init__(self, price):
        self._price = price

    @property
    def price(self):
        return self._price
    
    @price.setter
    def price(self, new_price):
        if new_price > 0 and isinstance(new_price, float):
            self._price = new_price
        else:
            print("Please enter a valid price")

    @price.deleter
    def price(self):
        del self._price
```

```python
penhouse = House(500)
```

```python
penhouse.price
```

```python
penhouse.price = 13.0
penhouse.price
```

```python
del penhouse.price
penhouse.price
```

# Protected methods

```python
class B:
    
    def __init__(self, attr):
        self.attr = attr
        
    def _protected_method(self):
        return 'protected'
        
    def __private_method(self):
        return 'private'
    
    def access_private_method(self):
        return self.__private_method()
```

```python
b = B('hi')
```

```python
b._protected_method()
```

```python
b.__private_method()
```

```python
b.access_private_method()
```
