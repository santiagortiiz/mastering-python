```python
from abc import ABC, abstractmethod
```

```python
class Figure_2D(ABC):
    
    @abstractmethod
    def area(self):
        pass
```

```python
# Cannot be instantiated
figure = Figure_2D()
```

```python
# area method must exists
class InvalidSquare(Figure_2D):
    
    def paint(self):
        print('filling figure')

invalid_square = InvalidSquare()
```

```python
class Square(Figure_2D):
    
    def area(self):
        print('calculating area')
```

```python
my_square = Square()
my_square.area()
```

```python
from abc import ABC, abstractmethod

class MyAbstractClass(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def speak(self):
        pass

class MyClass(MyAbstractClass):
    def speak(self):
        print(f"My name is '{self.name}'")

my_class = MyClass("Santiago")
my_class.speak()
```
