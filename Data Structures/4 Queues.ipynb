# 1 Queues based in lists

## Features
- FIFO: First in, first out
- Rear: Last item
- Front: First item
- Priority Queue: based on priority



```python
class ListQueue:

    def __init__(self):
        self.items = []
        self.size = 0

    def enqueue(self,data):
        self.items.insert(0, data)
        self.size += 1

    def dequeue(self):
        data = self.items.pop()
        self.size -= 1
        return data

    def traverse(self):
        total_items = self.size

        for item in range(total_items):
            print(self.items[item])
```

```python
food = ListQueue()
```

```python
food.enqueue('egg')
food.enqueue('ham')
food.enqueue('spam')
```

```python
food.traverse()
```

```python
food.dequeue()
```

# Queues based in stacks

```python
class Queue:
    
    def __init__(self) -> None:
        self.inbound_stack = []
        self.outbound_stack = []

    def enqueue(self, data):
        self.inbound_stack.append(data)

    def dequeue(self):
        if not self.outbound_stack:
            while self.inbound_stack:
                self.outbound_stack.append(self.inbound_stack.pop())

        return self.outbound_stack.pop() if self.outbound_stack else None
```

```python
numbers = Queue()
```

```python
numbers.enqueue(5)
numbers.enqueue(6)
numbers.enqueue(7)
```

```python
print(numbers.inbound_stack)
print(numbers.dequeue())
```

```python
numbers.enqueue(10)
print(numbers.inbound_stack)
print(numbers.outbound_stack)
```

```python
print(numbers.dequeue())
print(numbers.outbound_stack)
```

```python
print(numbers.dequeue())
print(numbers.outbound_stack)
print(numbers.inbound_stack)
```

# 3 Queues based in Nodes

```python
class TwoWayNode:
    
    def __init__(self, data, previous=None, next=None):
        self.data = data
        self.previous = previous
        self.next = next
```

```python
class Queue:
    
    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0

    def enqueue(self, data):
        new_node = TwoWayNode(data, None, None)
        
        if self.head is None:
            self.head = new_node
            self.tail = self.head
        else:
            new_node.previous = self.tail
            self.tail.next = new_node
            self.tail = new_node
        
        self.count += 1
    
    def dequeue(self):
        current = self.head

        if self.count == 1:
            self.count -= 1
            self.head = None
            self.tail = None

        elif self.count > 1:
            self.head = self.head.next
            self.head.previous = None
            self.count -= 1

        return current.data
        
```

```python
food = Queue()
```

```python
food.enqueue('eggs')
food.enqueue('ham')
food.enqueue('spam')
```
