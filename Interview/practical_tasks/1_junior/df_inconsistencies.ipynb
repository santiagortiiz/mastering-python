Design a function to detect name inconsistencies across assets in 2 environments of a remote system.
An asset has the following properties on each environment:

- name: str
- asset_id: int
- uploaded_date: date

The API of the system X has the method list_assets() that can be used to retrieve a dataframe with all assets in one environment:

from system import system_client

client = system_client(DEV | PROD)
client.list_assets()

```python
import pandas as pd

from system import system_client
```

```python
dev_client = system_client('DEV')
prod_client = system_client('PROD')
```

```python
dev_assets = dev_client.list_assets()
prod_assets = prod_client.list_assets()
```

```python
def get_df_inconsistencies(df_a: pd.DataFrame, df_b: pd.DataFrame, column_to_compare: str):
    df_1 = df_a.copy()
    df_2 = df_b.copy()

    # Identify what is in 1 but not in 2
    df_1['is_in_2'] = df_1[column_to_compare].isin(df_2[column_to_compare])
    in_1_not_2 = df_1.loc[df_1["is_in_2"] == False]

    # Identify what is in 2 but not in 1
    df_2['is_in_1'] = df_2[column_to_compare].isin(df_1[column_to_compare])
    in_2_not_1 = df_2.loc[df_2["is_in_1"] == False]
    return in_1_not_2, in_2_not_1
```

```python
in_1_not_2, in_2_not_1 = get_df_inconsistencies(dev_assets, prod_assets, 'name')
```
